```python
from ultralytics import SAM, YOLO
import torch

device = torch.device("cuda")
model_yolo = YOLO(r"D:\AOI-MKAC\MixPL\src\aoi_inspection\notebooks\best_VRS_Vetlom_Annotation.pt")
model_segment = SAM(r"D:\AOI-MKAC\MixPL\src\aoi_inspection\notebooks\sam2.1_s.pt")
```

```python
import cv2
path_image = r"D:\AOI-MKAC\MixPL\src\aoi_inspection\notebooks\3.jpg"

results = model_yolo(path_image)

boxes_xyxy = results[0].boxes.xyxy
confs = results[0].boxes.conf

for box, conf in zip(boxes_xyxy, confs):
    x1, y1, x2, y2 = box
    conf = float(conf)  # ép kiểu
    if conf >= 0.1:
        result = model_segment(
            path_image,
            bboxes=[x1, y1, x2, y2],
            conf=conf
        )
        mask = result[0].masks.data[0]
        pixel_count = len(mask)

        print(f"Bounding box [{x1:.0f}, {y1:.0f}, {x2:.0f}, {y2:.0f}] -> {int(pixel_count)} pixels")
result
```
